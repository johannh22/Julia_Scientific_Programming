```julia
using Plots # sets up to make the figures


# specify the given values; these will be subject to change later
r = 5       # specifies the radius
x0, y0 = 1, 2

function trialcircle(r, x0, y0)
    t = range(0., stop = 2Ï€, step = 0.005)
    x, y = r*cos.(t), r*sin.(t)
    x, y = x .+ x0, y .+ y0
    plot(x,y)
end
```

```julia
trialcircle(1, 1, 1)
```

```julia
function trialcircle!(r, x0, y0)
    t = range(0, stop = 2Ï€, step = 1/200)
    x, y = r*cos.(t), r*sin.(t)
    x, y = x .+ x0, y .+ y0
    plot!(x,y,
    grid = :false,
    legend = :false)
end
```

```julia
trialcircle(1, 0, 0)
trialcircle!(2, 0, 0)
```

```julia
trialcircle(0, 0, 0)


for r = .1:.1:4
    trialcircle!(r, 0, 0)
end
plot!()
```

```julia
plot!(grid = :false, legend = :false, aspect_ratio = 1)
```

```julia
trialcircle(0, 0, 0)

for r = .1:.1:4
    trialcircle!(r, r, 0)
end
plot!(grid = :false, legend = :false, aspect_ratio = 1)
```
